## 1. Линейная задача наименьших квадратов

Пусть функция $y = f(x)$ задана таблицей приближенных значений 

$y_i \approx f(x_i0, \quad i = 0, 1, \ldots n$,

полученных с ошибками $\varepsilon_i = y_i^0 - y_i$, где $y_i^0 = f(x_i)$. Если значения $y_i$ получены из эксперимента, то ошибки носят случайных характер и часто уровень погрешности ("шума" таблицы) бывает значительым.

Предположим, что для аппроксимации функции $f$ используется линейная модель:

$y = \Phi_m (x) = a_0 \varphi_0 (x) + a_1 \varphi_1 (x) + \ldots + a_m \varphi_m (x)$,

здесь $\varphi_0 (x), \varphi_1 (x), \ldots, \varphi_m (x)$ - заданные базисные функции;

$a_0, a_1, \ldots, a_m$ - параметры модели, являющиеся коэффициентами обобщенного многочлена $\Phi_m (x)$.

Одной из наиболее простых и часто используемых моделей является полиномиальная модель

$y = P_m (x) = a_0 + a_1 x + \ldots + a_m x^m$.

дописать сюда

Следует стремиться к тому, чтобы в этих точках выполнялись соответствующие приближеные равенства

$a_0 \varphi_0 (x_0) + a_1 \varphi_1 (x_0) + \ldots + a_m \varphi_m (x_0) \approx y_0$,

$a_0 \varphi_0 (x_1) + a_1 \varphi_1 (x_1) + \ldots + a_m \varphi_m (x_1) \approx y_1$,

$\ldots$

$a_0 \varphi_0 (x_n) + a_1 \varphi_1 (x_n) + \ldots + a_m \varphi_m (x_n) \approx y_n$.

В матричном виде: $Pa \approx y$.

Из различных критериев, позволяющих выбрать параметры $a_0, a_1, \ldots, a_m$, так, чтобы приближенные равенства удовлетворялись наилучшим образом, наиболее часто используется критерий наименьших квадратов.

Согласно этому критерию параметры выбираются так, что минимизировать среднеквадратичное уклонение

$\delta (\Phi_m, y) = \sqrt{\dfrac{1}{n+1} \sum\limits_{i=0}^{n} \big(\Phi_m (x_i)  - y_i\big)^2}$

обобщенного многочлена $\Phi_m (x) = \sum\limits_{j=0}^{m} a_j \varphi_j (x)$ от заданных табличных значений $y_i \quad (0 \le i \le n)$. Заметим, что миимум среднеквадратиного уклонения достигается при тех же значениях $a_0, a_1, \ldots, a_m$, что и минимум функции $s(a, y) = \sum\limits_{i=0}^{n} \bigg( \sum\limits_{j=0}^{m} a_j \varphi_j (x_i) - y_i \bigg)^2 = \Vert P a - y \Vert_2^2$,

причем

$\delta (\Phi_m, y)^2 = \dfrac{1}{n+1} s(a, y)$.

Итак, линейная задача метода наименьших квадратов состоит в следующем. Требуется найти (при фиксированном наборе функций $\varphi_0, \varphi_1, \ldots, \varphi_m$) обобщенный многочлен $\Phi_m^y (x)$, для которого среднеквадратичное уклонение принимает минимальное значение:

$\delta(\Phi_m^y, y) = \min\limits_{\Phi_m} \delta(\Phi_m, y)$.

Искомый обобщенный многочлен $\Phi_m^y$ будем далее называть многочленом наилучшего среднеквадратичного приближения.

## 2. Нормальная система

Существуют различные подходы к решению поставленной задачи. Простейший из них состоит в использовании необходимого экстремума функции $s$:

$\dfrac{\partial s}{\partial a_k} = 0, \quad k = 0, 1, \ldots, m$.


Вычисляя частные производные функции $a$ и изменяя порядок суммирования, от предыдущих равенств переходим к системе линейных алгебраических уравнений

$\sum\limits_{j=0}^{m} \bigg(\sum\limits_{i=0}^{n} \varphi_j (x_i) \varphi_k (x_i)\bigg) a_j = \sum\limits_{i=0}^{n} y_i \varphi_k (x_i), \quad k = 0, 1, \ldots, m$,

которая называется *нормальной системой метода наименьших квадратов*.

Как нетрудно видеть, нормальную систему можно записать в виде

$P^T P a = P^T y$,

или, используя матрица Грама $\Gamma = P^T P$ и вводя вектор $b = P^t y$, в виде

$\Gamma a  = b$.


#### Теорема

Пусть система функций $\varphi_0, \varphi_1, \ldots, \varphi_m$ линейно независима в точках $x_0, x_1, \ldots, x_n$. Тогда многочлен наилучшего среднеквадратичного приближения *существует и единственен*.

#### Замечание 1

Если $m = n$ и система функций $\varphi_0, \varphi_1, \ldots, \varphi_n$ линейно независима в точках $x_0, x_1, \ldots, x_n$, то многочлен $\Phi_n^y$, найденный методом наименьших квадратов, совпадает с интерполяционным многочленом $\Phi_n$. В самом деле, $\Phi_n(x_i) = y_i$ для всех $i =0, 1, \ldots, n$ и поэтому $\delta(\Phi_n, y) = 0$. Так как среднеквадратичное уклонение не может быть отрицательным, то $\Phi_n$ - многочлен наилучшего среднеквадратичного приближения. В силу его единственности $\Phi_n = \Phi_n^y$.

#### Замечание 2

Как правило, при использовании метода наименьших квадратов предполагается, что $m \ll n$. В этом случае метод обладает некоторыми сглаживающими свойствами.

Очень часто для приближения по методу наименьших квадратов используются алгебраические многочлены степени $m \le n$. Поскольку система функций $1, x, \ldots, x^m$ линейно независима в точках $x_0, x_1, \ldots, x_n$ при $m \le n$, в силу теоремы алгебраический многочлен наилучшего среднеквадратичного приближения существует и единственен.

Так как в случае приближения алгебраическими многочленами $\varphi_k (x) = x^k$,  то нормальная система принимает следующий вид:

$\sum\limits_{j=0}^{m} \bigg(\sum\limits_{i=0}^{n} x_i^{j+k}\bigg) a_j = \sum\limits_{i=0}^{n} y_i x_i^k, \quad k = 0, 1, \ldots, m$.

Запишем эту систему в развернутом виде в двух наиболее простых случаях $m=1$ и $m=2$.

#### 2) Используя функцию ```mnk``` найти многочлены $P_m, m = 0, 1, 2, \ldots$, по методу наименьших квадратов. Вычислить соответствующие им значения $\sigma_m$.

````python
def mnk(x, y, m, n):
    b = np.zeros_like(x)
    G = np.empty([m, m])
    for j in range(0, m+1, 1):
        for i in range(0, n+1, 1):
            b[j] = y[i]*x[i]**j
            for k in range(0, m+1, 1):
                G[j, k] = (x[i])**(k+j)
    a = np.linalg.solve(G, b)
    return a
````
